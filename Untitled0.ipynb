```python
import pandas as pd

df = pd.read_csv("world_population.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

top_10 = df.sort_values("2022 Population", ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top_10['Country/Territory'], top_10['2022 Population'], color='skyblue')
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Population (2022)")
plt.title("Top 10 Most Populated Countries (2022)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["2022 Population"] / 1e6, bins=30, color='coral')
plt.xlabel("Population (in millions)")
plt.ylabel("Number of Countries")
plt.title("Distribution of Population Across Countries (2022)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Create synthetic dataset
n = 1000  # Number of people
data = {
    'Age': np.random.normal(loc=35, scale=12, size=n).astype(int),  # Normal distribution of ages
    'Gender': np.random.choice(['Male', 'Female', 'Other'], size=n, p=[0.48, 0.48, 0.04])
}

df = pd.DataFrame(data)

# Clean unrealistic ages
df['Age'] = df['Age'].clip(lower=0, upper=90)

df.head()
```

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='pastel')
plt.title("Gender Distribution in Population")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title("Age Distribution in Population")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.grid(True)
plt.show()
```
